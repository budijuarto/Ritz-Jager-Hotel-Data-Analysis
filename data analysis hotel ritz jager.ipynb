from data above we can see that there are null data in column children, country_origin, agent, and company. We will fill children and country origin using mode. Agent and company using mean

```python
agentClean=df['agent'].mean()
companyClean=df['company'].mean()
df['agent']=df['agent'].fillna(agentClean)
df['company']=df['company'].fillna(companyClean)
df.head()
```

```python
childrenClean=df['children'].mode()[0]#You must add [0] that indicate you fill with the first value in index of mode
df['children']=df['children'].fillna(childrenClean)
df.head()
```

```python
cleanCountry=df['country_origin'].mode()[0] #You must add [0] that indicate you fill with the first value in index of mode
df['country_origin']=df['country_origin'].fillna(cleanCountry)
```

```python
df.isnull().sum()
```

Clean country code data into country name to easier interpretation

```python
import pycountry

#list the name country
list_alpha_2 = [i.alpha_2 for i in list(pycountry.countries)]
list_alpha_3 = [i.alpha_3 for i in list(pycountry.countries)]    

def country_flag(df):
    if (len(df['countrycode'])==2 and df['countrycode'] in list_alpha_2):
        return pycountry.countries.get(alpha_2=df['countrycode']).name
    elif (len(df['countrycode'])==3 and df['countrycode'] in list_alpha_3):
        return pycountry.countries.get(alpha_3=df['countrycode']).name
    else:
        return 'Invalid Code'
```

# 2. How much do guests pay for a room per night?

```python
df['adr'].describe()
```

```python
fig,ax = plt.subplots()
fig.set_size_inches(16, 8)

ax = sns.distplot(df.adr).set(title='Hotel Price Distribution')
```

as seen from above chart that there is outlier which minus price from hotel so we can ignore that. we can find the price of hotel per night by using the mean. But before that we must apply filter of type of hotel, price, cancellation,and adult and children must be at least one

```python
city_hotel = df[(df['hotel_type']=='City Hotel') & (df['is_canceled'] == 0) & (df.adr > 0) & (df.adults>0) & (df.children>0)]
resort_hotel = df[(df['hotel_type']=='Resort Hotel') & (df['is_canceled'] == 0) & (df.adr > 0) & (df.adults>0) & (df.children>0)]

print('For non-canceled reservation and all room types:\nCity Hotel mean price: ${}\nResort Hotel mean price: ${}'.format(city_hotel.adr.median(),resort_hotel.adr.median()))
```

```python
# changes reservation date into dates
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.dtypes
```

```python
#split year and month
monthPrice = df[['adr','reservation_status_date']]
avg=monthPrice.groupby([(monthPrice['reservation_status_date'].dt.year),(monthPrice['reservation_status_date'].dt.month)]).mean()
avg
```

```python
avgmonth=avg['adr'].pct_change()
plt.figure(figsize=(15,5))
plt.title('the price per night vary over the year')
plt.xlabel('Month and Year')
plt.ylabel('Percentage Change')
avgmonth.plot()
```

```python
no_cancel = df[df['is_canceled']==0]
pd.pivot_table(no_cancel, index=['reservation_status_date'], values='adr',aggfunc={'adr':np.mean}).plot(kind='line',color='r')
plt.title("The Price per Night Vary Over the Year by date")
plt.xlabel("Date")
plt.ylabel('Price')
```
